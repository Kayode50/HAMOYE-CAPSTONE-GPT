# **Modelling LOWER SEONDARY**

````python
from sklearn.linear_model import Ridge,RidgeCV,Lasso,LassoCV
from sklearn.metrics import mean_absolute_error,mean_squared_error,recall_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
import xgboost
from xgboost import XGBRegressor
````

````python

def data_split(data):

  x= data.drop(['Percentage'],axis = 1)
  y = data['Percentage']
  from sklearn.model_selection import train_test_split
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
  return x_train, x_test, y_train, y_test
````

````python
def minmax (x_train,x_test):
  from sklearn.preprocessing import MinMaxScaler,StandardScaler
  scaler =  MinMaxScaler()
  x_train = scaler.fit_transform(x_train)
  x_test = scaler.transform(x_test)
  return x_train,x_test
````

````python
def standard (x_train,x_test):
  from sklearn.preprocessing import MinMaxScaler,StandardScaler
  scaler =  StandardScaler()
  x_train = scaler.fit_transform(x_train)
  x_test = scaler.transform(x_test)
  return x_train,x_test
````

````python
from sklearn.linear_model import LinearRegression
````

**LINEAR REGRESSION**

````python
def linear_regression(x_train,y_train,x_test,y_test):
  lreg = LinearRegression() 
  lreg.fit(x_train,y_train)
  y_predp = lreg.predict(x_test)
  print('the linear_reg train score is {}'.format(lreg.score(x_train,y_train)))
  print('the linear_reg test score is {}'.format(lreg.score(x_test,y_test)))
  return lreg.score(x_test,y_test)
````

###Ridge Regression###


````python
def ridge_regression(x_train,y_train,x_test,y_test):
  ridge_regp = Ridge(alpha = 1 )
  ridge_regp.fit(x_train,y_train)
  y_predp = ridge_regp.predict(x_test)
  print('the ridge_reg train score is {}'.format(ridge_regp.score(x_train,y_train)))
  print('the ridge_reg test score is {}'.format(ridge_regp.score(x_test,y_test)))
  return ridge_regp.score(x_test,y_test)
````

````python
def ridge_cv(x_train,y_train,x_test,y_test):
  ridge_cvp = RidgeCV(alphas = [0.01,0.1,1,10])
  ridge_cvp.fit(x_test,y_test)
  print('the ridgecv train score is {}'.format(ridge_cvp.score(x_train,y_train)))
  print('the ridgecv test score is {}'.format(ridge_cvp.score(x_test,y_test)))
  return ridge_cvp.score(x_test,y_test)
````

###LASSO REGRESSION###

````python
def lasso_reg(x_train,y_train,x_test,y_test):
  lasso_regp = Lasso(alpha = 1) 
  lasso_regp.fit(x_train,y_train) 
  print('the lasso train score is {} '.format(lasso_regp.score(x_train,y_train))) 
  print('the lasso test score is {}'.format(lasso_regp.score(x_test,y_test))) 
  return lasso_regp.score(x_test,y_test)
````

**CROSS VALIDATION**

````python
def lassocv_reg(x_train,y_train,x_test,y_test):
  lasso_reg_cvp = LassoCV(alphas= [0.01,0.1,1,10])
  lasso_reg_cvp.fit(x_train,y_train)
  print('the lassocv train score is {} '.format(lasso_reg_cvp.score(x_train,y_train)))
  print('the lassocv test score is {}'.format(lasso_reg_cvp.score(x_test,y_test)))
  #LASSO PERFORMED BETTER
  return lasso_reg_cvp.score(x_test,y_test)
````

**Random Forest model**

````python
def random_forest(x_train,y_train,x_test,y_test):
  forest = RandomForestRegressor()
  forest.fit(x_train, y_train)
  forest_pred = forest.predict(x_test)
  print(f'The random forest root mean squared error is {mean_squared_error(y_test, forest_pred)**0.5}')
  print('the random forest train score is {}'.format(forest.score(x_train,y_train)))
  print('the random forest test score is {}'.format(forest.score(x_test,y_test)))
  return forest.score(x_test,y_test)
````

**Xgboost model**

````python

boost = XGBRegressor()
def xgboost_reg(x_train,y_train,x_test,y_test):
  #boost = XGBRegressor()
  global boost
  boost.fit(x_train, y_train)
  boost_pred = boost.predict(x_test)
  print(f'The xgboost root mean squared error is {mean_squared_error(y_test, boost_pred)**0.5}')
  print('the xgboost train score is {}'.format(boost.score(x_train,y_train)))
  print('the xgboost test score is {}'.format(boost.score(x_test,y_test)))
  return boost.score(x_test,y_test)
````

````python
def model(type,data_t,scaling):
  result =[type]
  
  x_train, x_test, y_train, y_test = data_split(data_t)
  if scaling == 'minmax':
    x_train, x_test = minmax(x_train, x_test)
  elif scaling == 'standard':
    x_train, x_test = standard(x_train, x_test)


  result.append(linear_regression(x_train,y_train,x_test,y_test))
  result.append(ridge_regression(x_train,y_train,x_test,y_test))
  result.append(ridge_cv(x_train,y_train,x_test,y_test))
  result.append(lasso_reg(x_train,y_train,x_test,y_test))
  result.append(lassocv_reg(x_train,y_train,x_test,y_test))
  result.append(random_forest(x_train,y_train,x_test,y_test))
  result.append(xgboost_reg(x_train,y_train,x_test,y_test))
  return result
````

````python
result1 = model('one hot with minmax',data_onehot,'minmax')
````

````python
result2 = model('one hot with standard',data_onehot,'standard')
````

````python
result3 = model('label enc with standard',data_labelenc,'minmax')
````

````python
result4 = model('label enc with standard',data_labelenc,'standard')
````
